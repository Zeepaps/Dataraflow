## Step-2 Finding a summer break

According to meteorologists, summer extends for the whole months of June, July, and August in the northern hemisphere and the whole months of December, January, and February in the southern hemisphere. So create a dataframe that holds just those months using the `datetime` index.

```python
#Since London falls under northern hemisphere, we'd be using (summer = london.loc['2023-06-01' : '2023-08-31']. Which is for June, July and August
summer = london.loc['2023-06-01' : '2023-08-31']
summer.head()
```

```python
#Getting the summary description for the summer data
summer[['tavg', 'prcp', 'tsun']].describe()
```

Next we find the perfect 2 weeks in these months to pick for the holiday

A good holiday week should have (Personal preference): A good weather with,

*   Moderate temperatures (not too hot or too cold)
*   Little or no rainfall
* Not too sunny (comfortable for exploring outdoors)





```python
good_weather = summer[
    (summer['tavg'].between(17, 22)) &
    (summer['prcp'] <= 1) &
    (summer['tsun'].between(150, 500))
]
good_weather.head()
```

```python
# Group by week and aggregate
weekly = summer.resample('W').agg({
    'tavg': 'mean',   # average temperature per week
    'prcp': 'sum',    # total rainfall per week
    'tsun': 'sum'     # total sunshine per week
})

# Display the first few weeks
print(weekly.head())
```

```python
# Creating a score based on ideal weather conditions
weekly['score'] = 0
weekly.loc[weekly['tavg'].between(17, 22), 'score'] += 1
weekly.loc[weekly['prcp'] <= 5, 'score'] += 1
weekly.loc[weekly['tsun'].between(150, 500), 'score'] += 1
```

```python
#locating a consecutive two weeks (14days) that falls under the condition
rolling_2week = summer[['tavg', 'prcp', 'tsun']].rolling(window=14).mean()
```

```python
# now selecting the good 2 weeks with the highest scores (most ideal weather) by sorting the scores
good_2week = rolling_2week[
    (rolling_2week['tavg'].between(17, 22)) &
    (rolling_2week['prcp'] <= 1) &
    (rolling_2week['tsun'].between(150, 500))
]
```

```python
good_2week.dropna().plot(grid=True, figsize=(10,5))
print(good_2week.dropna().head())
```

### **Summary of Findings**

Based on the 2023 summer weather data for London, the analysis identified **three potential 2-week periods** that offer the most favorable conditions for a summer holiday.

These optimal windows were around:

* **June 14 – June 27**
* **June 20 – July 3**
* **July 6 – July 19**

These periods were chosen because they met the ideal vacation criteria:

* **Moderate average temperatures** (around **18–20°C**) — comfortable for outdoor exploration without excessive heat.
* **Low rainfall levels** (less than **1 mm per day on average**) — minimizing chances of rain interruptions.
* **Balanced sunshine duration** (roughly **300–450 minutes daily**) — bright but not overly intense sunlight.

Overall, **late June to mid-July** appears to offer the most consistent and pleasant weather for a summer break in London, making it the best window to plan a back-to-back two-week holiday.
