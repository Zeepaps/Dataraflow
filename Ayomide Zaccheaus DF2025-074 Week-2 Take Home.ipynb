- Create a base class `BankAccount` with `owner`, `balance`, and methods `deposit()` and `withdraw()`.
- Create a subclass `SavingsAccount` that adds an `interest_rate` and a method `add_interest()` which increases the balance.
- Demonstrate deposits, withdrawals, and interest application.

````python
#Base class
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"{self.owner} deposited {amount}. New balance: {self.balance}")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient funds!")
        elif amount <= 0:
            print("Withdrawal amount must be positive.")
        else:
            self.balance -= amount
            print(f"{self.owner} withdrew {amount}. New balance: {self.balance}")

# Subclass
class SavingsAccount(BankAccount):
    def __init__(self, owner, balance=0, interest_rate=0.02):
        super().__init__(owner, balance)
        self.interest_rate = interest_rate

    def add_interest(self):
        interest = self.balance * self.interest_rate
        self.balance += interest
        print(f"Interest of {interest} added. New balance: {self.balance}")


#Demonstration
account = SavingsAccount("Ayo", 5000, 0.05)  # 5% interest rate

account.deposit(500)    #Deposit money
account.withdraw(2000)   #Withdraw money
account.add_interest()  #Apply interest
````

- Create a class `CountDown` that takes a number `n` and counts down to 1.  
- Implement the iterator protocol (`__iter__()` and `__next__()` methods).  
- Use it in a `for` loop to print numbers in reverse order.

````python
class CountDown:
    def __init__(self, n):
        self.n = n   #starting number

    def __iter__(self):
        return self  

    def __next__(self):
        if self.n > 0:
            current = self.n
            self.n -= 1
            return current
        else:
            raise StopIteration  


# Usage in a for loop
cd = CountDown(5)

for num in cd:
    print(num)
````

- Create a custom iterator `FileLineReader` that takes a filename and reads one line at a time.  
- Implement `__iter__()` and `__next__()` so it can be used in a `for` loop.  
- Test it by reading from a sample `.txt` file and printing each line.

````python
class FileLineReader:
    def __init__(self, filename):
        self.file = open(filename, "r")   

    def __iter__(self):
        return self

    def __next__(self):
        line = self.file.readline()
        if line:   #if line is not empty
            return line.strip()   #remove newline characters
        else:
            self.file.close()   #close file when done
            raise StopIteration


#Creating a sample text file for testing
with open("sample.txt", "w") as f:
    f.write("Hello, Zach!\n")
    f.write("This is week 2 at DataraFlow.\n")
    f.write("Hope you're learning 3.\n")

# Use FileLineReader
reader = FileLineReader("sample.txt")

for line in reader:
    print(line)
````

- Create an abstract class `Shape` with an abstract method `area()`.  
- Create subclasses `Rectangle` and `Circle` that implement the `area()` method.  
- Create a list of different shapes and loop through them, printing their area without checking their type.

````python
from abc import ABC, abstractmethod
import math

# Abstract base class
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

# Subclass: Rectangle
class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

# Subclass: Circle
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * (self.radius ** 2)

# Create objects
shapes = [
    Rectangle(5, 10),
    Circle(7),
    Rectangle(3, 8),
    Circle(2)
]

# Loop through without checking type
for shape in shapes:
    print("Area:", shape.area())
````

Create a class `Fibonacci` that generates Fibonacci numbers up to a given `n` using an iterator protocol.

````python
class Fibonacci:
    def __init__(self, n):
        self.n = n          #limit
        self.a, self.b = 0, 1
        self.count = 0      #track numbers generated

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < self.n:
            result = self.a
            self.a, self.b = self.b, self.a + self.b  #update sequence
            self.count += 1
            return result
        else:
            raise StopIteration  #stop iteration when done


for num in Fibonacci(10):
    print(num, end=" ")
````

# Library Management System  
  
## Scenario  
You are building a small library system to manage books, members, and borrowing operations. The system should support multiple types of books, track borrowed books, and allow iteration over members and books. The system should also demonstrate proper use of scopes and encapsulation.  
  
## Requirements and Concepts to Cover  
  
### 1. OOP & Inheritance  
- Create a base class `Book` with attributes: `title`, `author`, `year`.  
- Create subclasses:  
  - `EBook` with `file_size` and `format`.

  - `TeacherMember` with `department`.  
  
### 2. Polymorphism  
- Define a method `get_description()` in `Book` and override it in both `EBook` and `PrintedBook`.

### 3. Iterators

### 4. Scope  
- Use private and protected attributes to demonstrate scope:

### 5. Extra Features  
- Track which books are available and which are borrowed.  
- Allow searching books by title or author.  
- Count the total number of books borrowed by all members (use a class-level variable).  

````python
# ----------------------------
# Library Management System
# ----------------------------

# Base Book class
class Book:
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = year
        self.is_borrowed = False   # Track availability

    def get_description(self):
        return f"{self.title} by {self.author}, {self.year}"


# Subclasses of Book
class EBook(Book):
    def __init__(self, title, author, year, file_size, file_format):
        super().__init__(title, author, year)
        self.file_size = file_size
        self.file_format = file_format

    def get_description(self):
        return f"E-Book: {self.title} by {self.author}, {self.year}, Size: {self.file_size}MB, Format: {self.file_format}"


class PrintedBook(Book):
    def __init__(self, title, author, year, weight, dimensions):
        super().__init__(title, author, year)
        self.weight = weight
        self.dimensions = dimensions

    def get_description(self):
        return f"Printed Book: {self.title} by {self.author}, {self.year}, Weight: {self.weight}kg, Size: {self.dimensions}"


# Base Member class
class Member:
    total_borrowed_books = 0  # Class-level variable

    def __init__(self, name, member_id):
        self.name = name
        self.member_id = member_id
        self._borrowed_books = []  # Protected

    def borrow_book(self, book):
        # Default borrow behavior (to be overridden)
        if not book.is_borrowed:
            self._borrowed_books.append(book)
            book.is_borrowed = True
            Member.total_borrowed_books += 1
            print(f"{self.name} borrowed {book.title}")
        else:
            print(f"{book.title} is already borrowed.")

    def __iter__(self):
        # Iterator over borrowed books
        return iter(self._borrowed_books)


# Subclasses of Member
class StudentMember(Member):
    def __init__(self, name, member_id, grade_level):
        super().__init__(name, member_id)
        self.grade_level = grade_level

    def borrow_book(self, book):
        if len(self._borrowed_books) >= 3:
            print(f"{self.name} has reached the limit (3 books).")
        else:
            super().borrow_book(book)


class TeacherMember(Member):
    def __init__(self, name, member_id, department):
        super().__init__(name, member_id)
        self.department = department

    def borrow_book(self, book):
        if len(self._borrowed_books) >= 5:
            print(f"{self.name} has reached the limit (5 books).")
        else:
            super().borrow_book(book)


# Library Class
class Library:
    def __init__(self, library_name):
        self.__library_name = library_name   # Private attribute
        self.books = []

    # Getter & Setter for library name
    def get_library_name(self):
        return self.__library_name

    def set_library_name(self, new_name):
        self.__library_name = new_name

    def add_book(self, book):
        self.books.append(book)

    def search_by_title(self, title):
        return [book for book in self.books if title.lower() in book.title.lower()]

    def search_by_author(self, author):
        return [book for book in self.books if author.lower() in book.author.lower()]

    def available_books(self):
        return [book for book in self.books if not book.is_borrowed]

    def borrowed_books(self):
        return [book for book in self.books if book.is_borrowed]

    def __iter__(self):
        # Iterator over all books in library
        return iter(self.books)



if __name__ == "__main__":
    # Create library
    lib = Library("City Library")

    # Add books
    b1 = EBook("Python 101", "John Doe", 2020, 5, "PDF")
    b2 = PrintedBook("Deep Learning", "Jane Smith", 2018, 1.2, "8x11 inches")
    b3 = PrintedBook("Data Science", "Alan Turing", 2019, 0.8, "7x10 inches")
    lib.add_book(b1)
    lib.add_book(b2)
    lib.add_book(b3)

    # Create members
    s1 = StudentMember("Alice", 101, "Grade 10")
    t1 = TeacherMember("Mr. Bob", 201, "Computer Science")

    # Borrow books
    s1.borrow_book(b1)
    s1.borrow_book(b2)
    s1.borrow_book(b3)
    s1.borrow_book(b1)  # exceed student limit

    t1.borrow_book(b3)  # already borrowed by Alice
    t1.borrow_book(b2)

    # Iterate over borrowed books by Alice
    print("\nAlice's borrowed books:")
    for bk in s1:
        print(bk.get_description())

    # Iterate over library books
    print("\nAll books in library:")
    for bk in lib:
        print(bk.get_description())

    # Search by author
    print("\nBooks by 'Jane':")
    for bk in lib.search_by_author("Jane"):
        print(bk.get_description())

    # Show available books
    print("\nAvailable books:")
    for bk in lib.available_books():
        print(bk.get_description())

    # Show borrowed books
    print("\nBorrowed books:")
    for bk in lib.borrowed_books():
        print(bk.get_description())

    # Total borrowed books count
    print("\nTotal borrowed books across all members:", Member.total_borrowed_books)
````
