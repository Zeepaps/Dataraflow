```python
#Question 2
import random

score = 0
operators = ["+", "-", "*", "/"]

for i in range(5):
    num1 = random.randint(1, 10)
    num2 = random.randint(1, 10)
    op = random.choice(operators)

    # Ensure division produces integer answers
    if op == "/":
        num1 = num1 * num2  

    question = f"{num1} {op} {num2}"
    answer = eval(question)

    print(f"Question {i+1}: {question}")
    user_answer = float(input("Your answer: "))

    if user_answer == answer:
        print("Correct!\n")
        score += 1
    else:
        print(f"Wrong! The correct answer is {answer}\n")

print(f"Quiz Over! You scored {score}/5.")
```

```python
#Question 4
import datetime

#Asking the user for their birthday date and month 
day = int(input("Enter your birth day (DD): "))
month = int(input("Enter your birth month (MM): "))

today = datetime.date.today()
current_year = today.year

#Creating a date object for the next birthday
next_birthday = datetime.date(current_year, month, day)

#If the birthday already happened this year, we set it for next year
if next_birthday < today:
    next_birthday = datetime.date(current_year + 1, month, day)

#Calculate difference
days_left = (next_birthday - today).days

print(f"Your next birthday is in {days_left} days!")
```

1. What is the difference between a Python module and a package?

A module in Python is simply a single `.py` file that contains code such as functions, classes, or variables which can be imported and reused in other programs. A package, on the other hand, is a collection of related modules grouped together inside a directory, and it usually contains a special `__init__.py` file to indicate that the directory is a package. In simple terms, a module is one file, while a package is a folder containing multiple modules.

8. Explain the difference between `import module` and `from module import function`.

import module - This imports the whole module. To use anything inside it, you must prefix it with the module name.
example, 
import math
print(math.sqrt(16))

While, from module import function - imports only a specific function (or variable/class) from the module. It can be used directly without the module name.
example, 
from math import sqrt
print(sqrt(16))   


```python
#9. Write a program that asks the user for two numbers and prints their greatest common divisor (GCD). 
import math

num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

# Find GCD
gcd = math.gcd(num1, num2)

print(f"The GCD of {num1} and {num2} is {gcd}")
```

10. What is a virtual environment in Python, and why is it useful?

A virtual environment in Python is an isolated workspace that lets you install and manage dependencies (packages, libraries) for a specific project without affecting your global Python setup.

It is useful because different projects may require different versions of the same package, and a virtual environment ensures they don’t conflict. This makes projects cleaner and easier to manage.
    

```python
#11. Write a program that calculates the number of days between two given dates.

from datetime import datetime

# Asking the user for two dates in the format YYYY-MM-DD
date1 = input("Enter the first date (YYYY-MM-DD): ")
date2 = input("Enter the second date (YYYY-MM-DD): ")

# Converting the input strings into datetime objects
d1 = datetime.strptime(date1, "%Y-%m-%d")
d2 = datetime.strptime(date2, "%Y-%m-%d")

# Calculate the difference
difference = abs((d2 - d1).days)

print(f"The number of days between {date1} and {date2} is {difference} days.")
```
